```python
#Home Ownership
plt.figure(figsize=(10, 6))

# --- Prepare DataFrame for plotting ---
# 1. Create a copy to avoid modifying the original 'df' for temporary processing
df_plot_home_ownership = df.copy()

# 2. Filter out any NaN values from 'home_ownership' for this plot.
#    This addresses potential 'TypeError' if NaNs are present during sorting.
df_plot_home_ownership = df_plot_home_ownership.dropna(subset=['home_ownership'])

# 3. Apply the 'is_default_str' fix for the hue variable (as we did previously)
df_plot_home_ownership['is_default_str'] = df_plot_home_ownership['is_default'].astype(str)
print(f"df_plot_home_ownership['is_default_str'] dtype: {df_plot_home_ownership['is_default_str'].dtype}")

# --- Define the order for 'home_ownership' categories ---
# It's good practice to have a logical order.
# This list tries to put common categories first, and 'Unknown' (if imputed) last.
# 'ANY', 'OTHER', 'NONE' might be rare and can be at the end.
preferred_order = ['MORTGAGE', 'RENT', 'OWN', 'ANY', 'OTHER', 'NONE']

# Get existing unique categories in the data after dropping NaNs
existing_categories = df_plot_home_ownership['home_ownership'].unique()

# Construct the final order for the plot:
# Start with preferred categories that actually exist in the data
plot_order = [cat for cat in preferred_order if cat in existing_categories]

# Add any other categories not in preferred_order (e.g., 'UNKNOWN' from imputation)
# and sort them alphabetically, placing 'Unknown' last.
other_categories = [cat for cat in existing_categories if cat not in preferred_order]
if 'Unknown' in other_categories:
    other_categories.remove('Unknown')
    other_categories.sort()
    plot_order.extend(other_categories)
    plot_order.append('Unknown') # Always put 'Unknown' last if it exists
else:
    other_categories.sort()
    plot_order.extend(other_categories)

# Ensure the plot order is not empty
if not plot_order:
    print("Warning: No valid 'home_ownership' categories found to plot after filtering.")
else:
    sns.countplot(
        x='home_ownership',
        hue='is_default_str', # Use the string version for hue
        data=df_plot_home_ownership,
        order=plot_order, # Use the defined and filtered order
        palette='coolwarm'
    )
    plt.title('Home Ownership Distribution by Default Status')
    plt.xlabel('Home Ownership')
    plt.ylabel('Number of Loans')
    plt.xticks(rotation=45, ha='right') # Rotate labels for readability
    plt.tight_layout() # Adjust layout to prevent labels from overlapping

    # Customize the legend labels for clarity
    ax = plt.gca() # Get the current axes
    handles, labels = ax.get_legend_handles_labels()
    # Assuming '0' and '1' are the string labels from 'is_default_str'
    ax.legend(handles=handles, labels=['No Default', 'Default'], title='Default Status')

    plt.show()
```

```python
# --- Step 5.1: Handling Missing Values (Execution) ---

print("--- Applying Missing Value Imputation Strategies ---")

# 1. mort_acc (9.54% missing)
# Data Type: Numerical (count of mortgage accounts).
# Strategy: Impute with 0. It's plausible that a missing value here means the borrower has no mortgage accounts.
if 'mort_acc' in df.columns and df['mort_acc'].isnull().any():
    df['mort_acc'] = df['mort_acc'].fillna(0)
    print("Filled 'mort_acc' NaNs with 0 (assuming missing means no mortgage accounts).")
else:
    print("'mort_acc' column not found or has no missing values. Skipping imputation.")

# 2. emp_title (5.79% missing)
# Data Type: Categorical (textual, very high cardinality).
# Strategy: Impute with 'Missing'. Given it's a job title, mode imputation might not be very meaningful,
# and creating a new category for missing values preserves the information that the title was unknown.
if 'emp_title' in df.columns and df['emp_title'].isnull().any():
    df['emp_title'] = df['emp_title'].fillna('Missing')
    print("Filled 'emp_title' NaNs with 'Missing' category.")
else:
    print("'emp_title' column not found or has no missing values. Skipping imputation.")

# 3. emp_length (4.62% missing)
# Data Type: Categorical (ordinal, e.g., '< 1 year', '10+ years').
# Strategy: Impute with the mode. For ordinal categorical data, the most frequent value is a common choice.
# Alternatively, you could use 'Missing' as a new category if you believe the absence of data is significant.
if 'emp_length' in df.columns and df['emp_length'].isnull().any():
    # Calculate mode, ensuring to get the first one if there are multiple modes
    emp_length_mode = df['emp_length'].mode()[0]
    df['emp_length'] = df['emp_length'].fillna(emp_length_mode)
    print(f"Filled 'emp_length' NaNs with mode: '{emp_length_mode}'.")
else:
    print("'emp_length' column not found or has no missing values. Skipping imputation.")


# 4. title (0.44% missing)
# Data Type: Categorical (textual, loan titles).
# Strategy: Impute with 'Missing'. Similar to 'emp_title', a new category is robust.
# This column is often very similar to 'purpose' and might be dropped later if redundant.
if 'title' in df.columns and df['title'].isnull().any():
    df['title'] = df['title'].fillna('Missing')
    print("Filled 'title' NaNs with 'Missing' category.")
else:
    print("'title' column not found or has no missing values. Skipping imputation.")


# 5. pub_rec_bankruptcies (0.14% missing)
# Data Type: Numerical (count of bankruptcies).
# Strategy: Impute with 0. Similar to 'mort_acc', a missing count often implies 'zero' occurrences.
if 'pub_rec_bankruptcies' in df.columns and df['pub_rec_bankruptcies'].isnull().any():
    df['pub_rec_bankruptcies'] = df['pub_rec_bankruptcies'].fillna(0)
    print("Filled 'pub_rec_bankruptcies' NaNs with 0.")
else:
    print("'pub_rec_bankruptcies' column not found or has no missing values. Skipping imputation.")


# 6. revol_util (0.07% missing)
# Data Type: Numerical (revolving line utilization rate, percentage).
# Strategy: Impute with the median. This is a very small percentage of missing data, and median is robust to outliers.
if 'revol_util' in df.columns and df['revol_util'].isnull().any():
    revol_util_median = df['revol_util'].median()
    df['revol_util'] = df['revol_util'].fillna(revol_util_median)
    print(f"Filled 'revol_util' NaNs with median: {revol_util_median:.2f}.")
else:
    print("'revol_util' column not found or has no missing values. Skipping imputation.")

# --- Final check for missing values ---
print("\n--- Final Missing Values Check ---")
final_missing = df.isnull().sum()
final_missing_df = pd.DataFrame({'Missing Count': final_missing, 'Missing Percentage': (final_missing / len(df)) * 100})
final_missing_df = final_missing_df[final_missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False)
if final_missing_df.empty:
    print("All missing values have been handled! Your DataFrame is now clean.")
else:
    print("Remaining missing values (investigate further):")
    print(final_missing_df)
```

```python
# --- IMPORTANT: Ensure 'df' is loaded and initial feature engineering (like 'is_default', date features) is done BEFORE this block ---
# If you just restarted your kernel, you need to re-run your initial data loading and processing steps up to this point.


print("\n--- Identifying and Removing Problematic Columns for Encoding ---")

# 1. Drop 'loan_status': It's the original target, 'is_default' is the processed target.
if 'loan_status' in df.columns:
    df = df.drop(columns=['loan_status'])
    print("Dropped 'loan_status' as 'is_default' is the intended binary target.")
else:
    print("'loan_status' not found, assuming it was already handled or never existed.")

# 2. Drop 'address': High cardinality, will cause memory errors if One-Hot Encoded.
if 'address' in df.columns:
    df = df.drop(columns=['address'])
    print("Dropped 'address' due to very high cardinality and memory constraints for One-Hot Encoding.")
else:
    print("'address' not found, assuming it was already handled or never existed.")

# 3. Drop 'emp_title' and 'title': These were the first identified high-cardinality culprits.
columns_to_drop_for_encoding_final = ['emp_title', 'title']
for col in columns_to_drop_for_encoding_final:
    if col in df.columns:
        df = df.drop(columns=[col])
        print(f"Dropped '{col}' due to high cardinality or redundancy.")
    else:
        print(f"'{col}' not found, skipping drop (already dropped or never existed).")

print(f"\nDataFrame shape AFTER dropping problematic columns (before final encoding): {df.shape}")


print("\n--- Step 5.2: Encoding Remaining Categorical Features (One-Hot Encoding with sparse=True) ---")

# Identify all non-numerical columns that are not the target variable
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Ensure 'is_default' is explicitly not in this list (it should be numerical 0/1)
if 'is_default' in categorical_cols:
    categorical_cols.remove('is_default')
    print("Removed 'is_default' from categorical list (it should be your numerical target).")


print(f"Remaining categorical columns to be One-Hot Encoded: {categorical_cols}")

if not categorical_cols:
    print("No remaining categorical columns found for One-Hot Encoding. (This would indicate success if columns were handled!)")
else:
    # Apply One-Hot Encoding with sparse=True for memory efficiency
    df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int, sparse=True)

    # Replace the original df with the new encoded df
    df = df_encoded

    print(f"\nDataFrame shape AFTER One-Hot Encoding with sparse=True: {df.shape}")
    print("\nSample of new columns created (first 10):")
    # This filter aims to show some of the new columns by looking for pattern like original_colname_category
    new_encoded_cols_sample = [col for col in df.columns if any(cat_col + '_' in col for cat_col in categorical_cols)]
    print(new_encoded_cols_sample[:10])

    # Verify column types (should show 'Sparse[uint8, 0]' or similar for new columns)
    print("\nData types after encoding (sample):")
    print(df.dtypes.value_counts())

print("\nOne-Hot Encoding of remaining categorical features complete using sparse format.")
print("This should finally resolve the memory error.")
```

```python
from sklearn.preprocessing import StandardScaler

print("\n--- Step 5.3: Scaling Numerical Features (StandardScaler) ---")

# Identify numerical columns to scale
# We need to exclude the target variable 'is_default' and any sparse/one-hot encoded columns.
# One-hot encoded columns are typically int or Sparse[int] type.
# We want to scale columns that are floats or integers that represent continuous values.

# First, get all numerical column names (this will include sparse columns and the target)
all_numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

numerical_cols_to_scale = []
for col in all_numerical_cols:
    # Exclude the target variable 'is_default'
    if col == 'is_default':
        continue
    
    # Exclude columns that are sparse (these are your one-hot encoded columns)
    if isinstance(df[col].dtype, pd.SparseDtype):
        continue
        
    # Exclude columns that are already binary (0/1) but not sparse (e.g., if any non-sparse dummy vars were created or existed)
    # This check is a bit redundant if you've done sparse=True for all dummies, but good for robustness.
    if df[col].nunique() == 2 and df[col].min() == 0 and df[col].max() == 1:
        continue
    
    # If it passes all checks, it's a true continuous/numerical feature that needs scaling
    numerical_cols_to_scale.append(col)

print(f"Numerical columns selected for scaling: {numerical_cols_to_scale}")
if not numerical_cols_to_scale:
    print("Warning: No numerical columns found that need scaling. This might be unexpected if you have features like 'loan_amnt', 'annual_inc', etc.")

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply scaling to the selected columns
# Use .values to work with the underlying numpy array, then convert back to DataFrame
# Ensure you are not trying to scale a column that contains NaN, although we filled all of them.
df[numerical_cols_to_scale] = scaler.fit_transform(df[numerical_cols_to_scale])


print("\nNumerical features scaled successfully.")
print("Sample of scaled data (first 5 rows of scaled columns):")
if numerical_cols_to_scale: # Only print if there are columns to show
    print(df[numerical_cols_to_scale].head())
else:
    print("No numerical columns were scaled.")

# Check descriptive statistics after scaling (mean should be close to 0, std close to 1)
print("\nDescriptive statistics of scaled numerical features (sample):")
if numerical_cols_to_scale: # Only print if there are columns to show
    print(df[numerical_cols_to_scale].describe().T[['mean', 'std']].head()) # Transpose for better readability
else:
    print("No numerical columns to show statistics for.")
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

print("\n--- Step 6.1: Splitting Data into Training and Testing Sets ---")

# Define features (X) and target (y)
# X will be all columns EXCEPT 'is_default'
X = df.drop('is_default', axis=1)

# y will be the 'is_default' column
y = df['is_default']

# Split the data into training and testing sets
# test_size=0.20 means 20% of data will be used for testing, 80% for training
# random_state for reproducibility (you'll get the same split every time)
# stratify=y is important for classification problems to ensure
# that the proportion of target classes (0s and 1s) is roughly the same
# in both training and testing sets as it is in the original dataset.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nData split into training and testing sets successfully.")
```

```python
#Logistic Regression Model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print("\n--- Step 6.2: Training Logistic Regression Model ---")

# Initialize the Logistic Regression model
# solver='liblinear' is a good default for relatively small datasets or L1/L2 regularization.
# random_state for reproducibility
log_model = LogisticRegression(solver='liblinear', random_state=42, n_jobs=-1) # n_jobs=-1 to use all available cores

# Train the model on the training data
print("Training Logistic Regression model...")
log_model.fit(X_train, y_train)
print("Logistic Regression model training complete.")
```

```python
#Random Forest Classifier Model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print("\n--- Step 6.2: Training Random Forest Classifier Model ---")

# Initialize the Random Forest Classifier
# n_estimators: number of trees in the forest (start with a moderate number, e.g., 100)
# random_state for reproducibility
# n_jobs=-1 to use all available cores, speeds up training
print("Initializing Random Forest Classifier...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Train the model on the training data
print("Training Random Forest Classifier model...")
rf_model.fit(X_train, y_train)
print("Random Forest Classifier model training complete.")
```
